```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
admissions_df = pd.read_csv('ADMISSIONS.csv')
icustays_df = pd.read_csv('ICUSTAYS.csv')
diagnoses_df = pd.read_csv('DIAGNOSES_ICD.csv')
patients_df = pd.read_csv('PATIENTS.csv')
procedures_df = pd.read_csv('PROCEDURES_ICD.csv')
diagnoses_icd_df = pd.read_csv('D_ICD_DIAGNOSES.csv')
caregivers_df = pd.read_csv('CAREGIVERS.csv')
```

```python
admissions_df['ADMITTIME'] = pd.to_datetime(admissions_df['ADMITTIME'])
```

```python
# Plot Histogram of ICU Length of Stay
plt.figure(figsize=(12, 6))
sns.histplot(icustays_df['LOS'], bins=30, kde=True, color= 'red')
plt.title('Distribution of ICU Length of Stay')
plt.xlabel('Length of Stay (days)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Convert both to string to ensure a successful merge
diagnoses_icd_df['ICD9_CODE'] = diagnoses_icd_df['ICD9_CODE'].astype(str)
diagnoses_df['ICD9_CODE'] = diagnoses_df['ICD9_CODE'].astype(str)

# Retry merging the datasets
merged_diagnoses_df = pd.merge(diagnoses_icd_df, diagnoses_df, on='ICD9_CODE', how='inner')

# Count occurrences of each diagnosis
top_diagnoses = merged_diagnoses_df['LONG_TITLE'].value_counts().head(10)

# Plot the most common diagnoses
plt.figure(figsize=(12, 6))
sns.barplot(x=top_diagnoses.values, y=top_diagnoses.index, color = 'yellow')
plt.title('Top 10 Most Common Diagnoses')
plt.xlabel('Frequency')
plt.ylabel('Diagnosis')
plt.grid(True)
plt.show()
```

```python
# Merge PATIENTS and ADMISSIONS datasets on SUBJECT_ID
merged_df = pd.merge(patients_df, admissions_df, on='SUBJECT_ID', how='inner')

# Ensure ADMITTIME is in datetime format
merged_df['ADMITTIME'] = pd.to_datetime(merged_df['ADMITTIME'])
merged_df['DOB'] = pd.to_datetime(merged_df['DOB'])
# Calculate age at admission
merged_df['AGE_AT_ADMISSION'] = (merged_df['ADMITTIME'].dt.year - merged_df['DOB'].dt.year).clip(lower=0)

# Define age groups
bins = [0, 20, 40, 60, 80, 100, 120]
labels = ['0-20', '21-40', '41-60', '61-80', '81-100', '100+']
merged_df['AGE_GROUP'] = pd.cut(merged_df['AGE_AT_ADMISSION'], bins=bins, labels=labels, right=False)

# Calculate mortality rate per age group using EXPIRE_FLAG
mortality_rate = merged_df.groupby('AGE_GROUP')['EXPIRE_FLAG'].mean() * 100

# Plot Mortality Rate by Age Group
plt.figure(figsize=(10, 6))
sns.barplot(x=mortality_rate.index, y=mortality_rate.values)
plt.title('Mortality Rate by Age Group (Using PATIENTS & ADMISSIONS Data)')
plt.xlabel('Age Group')
plt.ylabel('Mortality Rate (%)')
plt.ylim(0, 100)
plt.grid(True)
plt.show()
```

```python
# Count the number of caregivers per specialty
specialty_counts = caregivers_df['DESCRIPTION'].value_counts().head(10)

# Plot a horizontal bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=specialty_counts.values, y=specialty_counts.index, color='lightgreen')
plt.title('Caregivers and Their Specialties')
plt.xlabel('Count')
plt.ylabel('Specialty')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Admissions Over Time (Line Plot)
plt.figure(figsize=(12, 6))
admissions_df.set_index('ADMITTIME').resample('Y').size().plot(color='green')
plt.title('Number of Admissions Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Admissions')
plt.grid(True)
plt.show()
```
