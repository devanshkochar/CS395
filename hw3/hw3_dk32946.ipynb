```python
import pandas as pd
import spacy
import scispacy
import torch
import re
from transformers import AutoTokenizer, AutoModel
from gensim.models import Word2Vec
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from collections import Counter
from tqdm import tqdm
import numpy as np
```

```python
# Load MIMIC-III Dataset (Assuming CSV files are available)
noteevents_df = pd.read_csv('NOTEEVENTS.csv', low_memory=False)
diagnoses_icd_df = pd.read_csv('DIAGNOSES_ICD.csv', low_memory=False)

```

```python
# Filter for Heart Failure (Example Disease)
heart_failure_icd = ['4280']  # ICD-9 codes for heart failure
heart_failure_df = diagnoses_icd_df[diagnoses_icd_df['ICD9_CODE'].isin(heart_failure_icd)]
```

```python
# Merge with Note Events
df = noteevents_df.merge(heart_failure_df, on=['SUBJECT_ID', 'HADM_ID'], how='inner')
df = df[df["CATEGORY"] == "Discharge summary"]
df=df.head(500)
notes = df['TEXT'].dropna().tolist()
```

```python
# Load NLP models
nlp_spacy = spacy.load('en_core_web_sm')
nlp_scispacy = spacy.load('en_ner_bc5cdr_md')
clinical_bert = AutoModel.from_pretrained("emilyalsentzer/Bio_ClinicalBERT")
clinical_tokenizer = AutoTokenizer.from_pretrained("emilyalsentzer/Bio_ClinicalBERT")

```

```python
def extract_entities(nlp, texts):
    entities = []
    
    # Debug: Check input format
    #print("Sample Input Texts:", texts[:2])  # Print first 2 entries to check
    
    for text in tqdm(texts, desc="Processing with NLP Model"):
        #print(f"\nProcessing text:\n{text[:200]}...")  # Print first 200 characters for debugging

        if not isinstance(text, str) or len(text.strip()) == 0:
            print("Skipping invalid or empty text:", text)
            continue  # Skip empty or invalid text
        
        doc = nlp(text)

        # Extract words (not punctuation/spaces)
        tokenized_text = [token.text for token in doc if not token.is_punct and not token.is_space]

        #print("Tokenized:", tokenized_text[:20])  # Print first 20 tokens for debugging

        if tokenized_text:  # Only append non-empty lists
            entities.append(tokenized_text)

    return entities

```

```python
def extract_entities_bert(texts):
    entities = []
    for text in tqdm(texts, desc="Processing with ClinicalBERT"):
        # Tokenize the text using ClinicalBERT tokenizer
        tokens = clinical_tokenizer(text, return_tensors="pt", truncation=True, max_length=512)
        
        # Extract tokenized words
        words = clinical_tokenizer.convert_ids_to_tokens(tokens["input_ids"][0])

        # Remove punctuation, numbers, spaces, and special characters
        filtered_words = [re.sub(r"[^a-zA-Z]", "", word) for word in words]  # Keep only letters

        # Remove single-character tokens (except "I" or "a" if needed)
        filtered_words = [word for word in filtered_words if len(word) > 1]

        entities.append(filtered_words)

    return entities
```

```python

spacy_entities = extract_entities(nlp_spacy, notes)
```

```python
scispacy_entities = extract_entities(nlp_scispacy, notes)
```

```python
bert_entities = extract_entities_bert(notes)
```

```python
def plot_tsne(model, title):
    words = list(model.wv.key_to_index.keys())
    vectors = np.array([model.wv[word] for word in words])  # Convert to NumPy array
    
    tsne = TSNE(n_components=2, perplexity=30, random_state=42)
    reduced_vectors = tsne.fit_transform(vectors)
    
    plt.figure(figsize=(12, 8))
    for i, word in enumerate(words[:100]):  # Plot top 100 words
        plt.scatter(reduced_vectors[i, 0], reduced_vectors[i, 1])
        plt.annotate(word, (reduced_vectors[i, 0], reduced_vectors[i, 1]))
    
    plt.title(f"t-SNE visualization of {title} Word2Vec embeddings")
    plt.show()
```

```python
# Generate separate t-SNE plots
plot_tsne(spacy_model, "SpaCy")
```

```python
# Generate separate t-SNE plots
plot_tsne(spacy_model, "SciSpaCy")
```

```python
plot_tsne(bert_model, "ClinicalBERT")
```
